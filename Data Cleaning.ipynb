```python
import pandas as pd
```

```python
df = pd.read_excel(r"/datalab/Customer Call List.xlsx")
df
```

```python
#REMOVING ANY DUPLICATS WE GOT IN THE DATAFRAME USING DF.DROP_DUPLICATE FUCTION
df = df.drop_duplicates()
df
```

```python
#WE REMOVED THE "Not_Useful_Column" BY USING THE DF.DROP FUNCTION
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#SELECTING LSTRIP AND RSTRIP TO REMOVE THE ERRORS
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
#CAN ASLO COMBINE BY DOING df["Last_Name"] = df["Last_Name"].str.strip("123..._/")
```

```python
#WE CLEANED UP THE nan ,NaN,N/A, USING THE FILLNA FUNCTION
df['Phone_Number'] = df["Phone_Number"].str.replace('N/a','')
df.Phone_Number.fillna("", inplace=True)
df


```

```python
#WE MADE 3 NEW COLLUM FROM THE STREET_ADDRESS
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2,  expand=True)
df
```

```python
#REPLACING THE YES TO Y AND NO TO N
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
#SAME AS THE ABOVE FOR THE DO_NOT_CONTACT COLLUM
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df.fillna("", inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

```python
#WE RESET THE INDEX USING THE DROP_INDEX FUNCTION
df = df.reset_index(drop = True)
df
```
